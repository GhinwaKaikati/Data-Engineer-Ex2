```python
import pandas as pd
import requests

df1 = pd.DataFrame()

for page in range(100, 106):
    url = f'https://unstats.un.org/SDGAPI/v1/sdg/Series/Data?page={page}'
    response = requests.get(url)
    data = response.json().get('data', [])

    page_df = pd.DataFrame(data)
    df1 = pd.concat([df1, page_df], ignore_index=True)

```

```python
import pandas as pd
import requests

df2 = pd.DataFrame()

for page in range(117430, 117436):
    url = f'https://unstats.un.org/SDGAPI/v1/sdg/Series/Data?page=117430={page}'
    response = requests.get(url)
    data = response.json().get('data', [])

    page_df = pd.DataFrame(data)
    df2 = pd.concat([df2, page_df], ignore_index=True)
```

```python
df2
```

```python
df3 = pd.concat([df1, df2], ignore_index=True)
```
